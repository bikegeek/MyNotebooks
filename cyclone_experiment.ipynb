```python
import numpy as np
import pandas as pd
import cartopy as cp
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
```

```python
def rescale_lonlist(lons):
    """ Convert lons in range -180 to 180 to 0 to 360 """
    
    rlons = []
    for curlon in lons:
        adjlon = curlon % 360
        rlons.append(adjlon)   

    return rlons
   
    
    
```

```python
def rescale_lon_to_180(lons:float) -> float:
        """! Rescales longitudes from 0 to 360 degrees to -180 to 180 degrees,
             using the same logic employed by MET
             Args:
             @param lons:  A list of longitudes as a float value from 0. to 360. degrees
             
             
             Returns:
             @param adj_lon: A list of "adjusted" longitudes, now ranging from -180. to +180.
             
        """
        rescaled_lons = []
        for cur_lon in lons:
            adj_lon = ((cur_lon + 180) % 360) - 180
            rescaled_lons.append(adj_lon)
#             print("lon to be rescaled: ",  cur_lon, "lon after rescaling 180: ", adj_lon)
           

        return rescaled_lons
```

```python
def sanitize_lonlist(lons):
    new_list = []
    oldval = 0
    treshold = 10  # used to compare adjacent longitudes
    for ix,ea in enumerate(lons):
        diff = oldval - ea
        if (ix>0):
            if (diff>treshold):
                orig = ea
                ea = ea+360
                if ea > 360:
                    ea = orig
#                 print("Sanitizing, orig= ", orig, " ea calculated: ", ea)
#                 print(" ea value used: ", ea)

                    
        oldval = ea
        new_list.append(ea)
        
    
    return new_list
```

```python
orig_xdateline = [141.2, 146.2, 149.3, 152.9, 156.2, 159.7, 162.9, 166.3, 168.8, 171.5, 174., 176.9, 179.8, -178.7, -177.4, 146., 149.3, 152.8,156.0, 159.5, 162.9, 165.8, 168.3, 170.9, 172.8, 175.8, 177.4,-180., -178.]
ydateline = [34.7, 35.7, 36.9, 37.6, 38.,38.6, 38.8,38.9, 39.,38.7, 38.1,38.2, 38.8,38.7, 37.8, 35.4, 36.9, 37.6, 38.1,38.3, 38.8,39.0, 38.9, 38.8, 38.7, 39.0, 39.1, 39.9, 40.1]
```

```python
# rescale lons from -180 to 180 to 0 to 360
xdateline_360 = rescale_lonlist(orig_xdateline)
```

```python
##
# Convert Original lons to 0-360
# Sanitize
# Plot lons in 0-? (some values may be >360 from sanitizing)
#
#
fig = plt.figure(figsize=(15,13))
prj = ccrs.PlateCarree()
cm_lon = 180
ax = plt.axes(projection=ccrs.PlateCarree(central_longitude=cm_lon))
ax.set_global() 
ax.coastlines(alpha=0.3)
ax.gridlines(draw_labels=False, xlocs=[180, -180])
gl = ax.gridlines(crs=prj,
                          draw_labels=True, linewidth=1, color='gray',
                          alpha=0.5, linestyle='--')

# Sanitize lons in 0-360 degree range.
san_xdateline = sanitize_lonlist(xdateline_360)
for idx, x in enumerate(san_xdateline):
    print("lon_360:", xdateline_360[idx], " sanitized lon: ", x)


plt.plot(san_xdateline, ydateline, transform=ccrs.PlateCarree(), color='green', marker='+') 
transform = ccrs.PlateCarree()._as_mpl_transform(ax)
index = 0 

for x,y in zip(san_xdateline, ydateline):
#     pt_text = "(" + str(x) + "," + str(y) + ")"
    pt_text = str(x)
    if index == 0:
       plt.annotate(pt_text, (x-6,y+4), xycoords=transform)
       index += 1
#     else:
#        plt.annotate(pt_text, (x-6,y+4), xycoords=transform)
#        index += 1

plt.title("RESCALE to 0-360,SANITIZE ALL LONS, KEEP in 0-360")
# plt.savefig("/Users/minnawin/Desktop/jupyter_rescale_san.png")
plt.show()
```

```python
##
# Rescale 0-360
# Sanitize
# Rescale back to -180 to 180
#
## 
fig = plt.figure(figsize=(15,13))
prj = ccrs.PlateCarree()
cm_lon = 180
ax = plt.axes(projection=ccrs.PlateCarree(central_longitude=cm_lon))
ax.set_global() 
ax.coastlines(alpha=0.3)
ax.gridlines(draw_labels=False, xlocs=[180, -180])
gl = ax.gridlines(crs=prj,
                          draw_labels=True, linewidth=1, color='gray',
                          alpha=0.5, linestyle='--')

# Sanitize lons in 0-360 degree range.
san_xdateline = sanitize_lonlist(xdateline_360)

# rescale the sanitized lons back to -180 to 180 degree range.
cleaned =  rescale_lon_to_180(san_xdateline)


plt.plot(cleaned, ydateline, transform=ccrs.PlateCarree(), color='red', marker='+') 
# plt.plot(cleaned, ydateline, transform=ccrs.PlateCarree(), color='red', marker='+') 
transform = ccrs.PlateCarree()._as_mpl_transform(ax)
index = 0 

for x,y in zip(cleaned, ydateline):
#     pt_text = "(" + str(x) + "," + str(y) + ")"
    pt_text = str(x)
    if index == 0:
       plt.annotate(pt_text, (x-6,y+4), xycoords=transform)
       index += 1
#     else:
#        plt.annotate(pt_text, (x-6,y+4), xycoords=transform)
#        index += 1

plt.title("RESCALE to 0-360, SANITIZE ALL LONS, RESCALED BACK TO -180-180")
plt.savefig("/Users/minnawin/Desktop/jupyter_rescale_san_rescale.png")
# plt.title("SANITIZE and CLEAN ALL LONS")
plt.show()
```
